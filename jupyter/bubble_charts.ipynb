````python
# Return Excel table back to DataFrame
wb = xw.books.active
sheet1 = wb.sheets[0]
# new_data = sheet1.range("A1").expand("table").value
df = xw.load(index=1, header=1)
````

````python
xw.view(df)
````

````python
# create data
data = [
    go.Scatter(
        x=df["horsepower"],
        y=df["mpg"],
        text=df["name"],
        mode="markers",
        marker=dict(size=2*df["cylinders"])
    )
]
````

````python
# create layout
layout = go.Layout(
    title="Bubble Chart"
)
````

````python
# create figure
fig = go.Figure(data=data, layout=layout)
````

````python
# Create plot in html
pyo.plot(fig, filename="bubble_chart.html")
````
